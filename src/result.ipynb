```python
#!/usr/bin/env python3
import matplotlib.pyplot as plt
import numpy as np
```

```python
# L o arquivo de utilizao com cabealho dinmico
with open('utilizacao_vs_clientes.txt') as f:
    header = f.readline().strip().split()
    nomes_dispositivos = header[1:]  # Ignora o 'N' inicial

dados = np.loadtxt('utilizacao_vs_clientes.txt', skiprows=1)
clientes = dados[:, 0]
utilizacoes = dados[:, 1:].T  # Transpe para facilitar o plot

plt.figure(figsize=(6, 4))
for idx, nome in enumerate(nomes_dispositivos):
    plt.plot(clientes, utilizacoes[idx], marker='o', label=nome)

plt.title('Utilizao dos Dispositivos de Acordo com a Quantidade de Clientes')
plt.xlabel('Nmero de Clientes')
plt.ylabel('Utilizao (%)')
plt.xticks(np.arange(int(min(clientes)), int(max(clientes)) + 1, 1))
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
